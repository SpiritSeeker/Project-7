```python
import numpy as np
import cv2
import face_recognition
```

```python
encs = np.zeros((1000,128))
```

```python
for i in range(1000):
    print(i)
    img = cv2.imread('face_data/sasank/'+str(i+1)+'.jpg')
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    boxes = face_recognition.face_locations(rgb, model='cnn')
    encs[i] = face_recognition.face_encodings(rgb, boxes)[0]
```
